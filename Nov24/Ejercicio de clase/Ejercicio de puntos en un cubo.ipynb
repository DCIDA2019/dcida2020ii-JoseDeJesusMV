Se puede observar una grafica de la proporcion que tiene el volumen de una bola con el volumen de un hipercubo
ambos de dimension D y de radio y lado R respectivamente

```python
import numpy as np
import math
from scipy.special import gamma, factorial
import scipy as sc
import matplotlib.pyplot as plt

Props = np.zeros(20,float)
Dims = np.zeros(20,float)

def proporcion(Rho,D):
    result = 2*np.power(np.pi,D/2)/(np.power(2,D-1)*sc.special.gamma(D/2)*D)
    return result
```

```python
R=50

for i in range(20):
    D=i+2
    Dims[i]=i+2
    Proporcion = proporcion(R,D)
    Props[i]=Proporcion
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(Dims,Props,'k-')
plt.xlabel('D',fontsize=16)
plt.xlim(2,22)
plt.ylabel('$Proporcion$',fontsize=16)
plt.title('Proporciones',fontsize=16)
plt.grid();
```
