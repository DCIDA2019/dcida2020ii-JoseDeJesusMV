```python
merged = pd.merge(df1, df2, how='outer',
                  left_on='Lon', right_on='Tmed')
merged
```

```python
#quitar las latitudes y longitudes
merged = merged.drop('Lon_x', 1)
merged = merged.drop('Lat_x', 1)
merged = merged.drop('Lon_y', 1)
merged = merged.drop('Lat_y', 1)
merged.head()
```
